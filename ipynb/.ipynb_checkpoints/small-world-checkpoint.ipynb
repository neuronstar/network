#### Tests

```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_nodes_from([2,3])
```

```python
G.add_edges_from([(1,2),(1,3)])
```

```python
print G.number_of_nodes(), G.number_of_edges()
```

```python
print G.nodes(), '\n', G.edges(), '\n', G.neighbors(1) 
```

```python
H=nx.DiGraph(G)
```

```python
nx.draw(G)
```

```python
nx.draw_random(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spectral(G)
```

```python
## Using graphviz
nx.draw_graphviz(G)
```
